```python
import pandas as pd 
import numpy as np
import pickle
import plotly.express as px
from imblearn.over_sampling import SMOTE
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df['Class'].value_counts()
```

```python
X = df.iloc[:,:-1]
y = df['Class']
```

```python
X.fillna(X.median())
```

```python
counts = df['Class'].value_counts()

fig = px.bar(
    x=counts.index,
    y=counts.values,
    color=counts.index,  # color by class
    color_discrete_map={0: 'skyblue', 1: 'salmon'},
    labels={'x': 'Class', 'y': 'Count'}
)

fig.show()
```

# Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
print("Accuracy AFTER SMOTE:", round(accuracy_score(y_test, y_pred_rf)*100, 2), "%\n")
print("Classification Report BEFORE SMOTE:\n", classification_report(y_test, y_pred_rf))

sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix BEFORE SMOTE")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# Random Forest after smoting

```python
smote = SMOTE(sampling_strategy='minority', random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)
```

```python
counts = y_train_sm.value_counts()
```

```python
fig = px.bar(
    x=counts.index,
    y=counts.values,
    color=counts.index,  # color by class
    color_discrete_map={0: 'skyblue', 1: 'red'},
    labels={'x': 'Class', 'y': 'Count'}
)

fig.show()
```

```python
rf = RandomForestClassifier(n_estimators=100)
```

```python
rf.fit(X_train_sm, y_train_sm)
y_pred_sm = rf.predict(X_test)
```

```python
print("Accuracy AFTER SMOTE:", round(accuracy_score(y_test, y_pred_sm)*100, 2), "%\n")
print("Classification Report AFTER SMOTE:\n", classification_report(y_test, y_pred_sm))

sns.heatmap(confusion_matrix(y_test, y_pred_sm), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix AFTER SMOTE")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# Random Forest with class weight

```python
rf = RandomForestClassifier(class_weight='balanced')
```

```python
rf.fit(X_train_sm, y_train_sm)
y_pred_sm = rf.predict(X_test)
```

```python
print("Accuracy AFTER SMOTE:", round(accuracy_score(y_test, y_pred_sm)*100, 2), "%\n")
print("Classification Report AFTER SMOTE:\n", classification_report(y_test, y_pred_sm))

sns.heatmap(confusion_matrix(y_test, y_pred_sm), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix AFTER SMOTE")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
with open("fraud_model.pkl", "wb") as f:
    pickle.dump(rf, f)
```
